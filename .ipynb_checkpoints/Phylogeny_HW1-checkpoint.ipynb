```python
# импортируем библиотеки
import matplotlib
import matplotlib.pyplot as plt
from Bio import Phylo
from io import StringIO
from urllib.request import urlopen

from ete3 import Tree, TreeStyle, NodeStyle
import random
```

```python
### Задания на Bio.Phylo ###
```

```python
# Задание 1. Подгружаем данные
target_url = 'https://www.jasondavies.com/tree-of-life/life.txt'
treedata = urlopen(target_url).read().decode('utf-8')
handle = StringIO(treedata)
tree = Phylo.read(handle, "newick")
```

```python
# Задание 2. Рисуем с помощью draw_ascii()
Phylo.draw_ascii(tree)
```

```python
# Задание 3. Рисуем с помощью draw()
plt.rcParams['figure.figsize'] = [100, 100]
Phylo.draw(tree)
```

```python
# Задание 4. Сохраняем в png svg, pdf
plt.rcParams['figure.figsize'] = [100, 100]
Phylo.draw(tree, do_show=False)
plt.savefig("my_tree_1.png", format='png')
plt.savefig("my_tree_1.svg", format='svg')
plt.savefig("my_tree_1.pdf", format='pdf')
```

```python
# Задание 5. Переводим в формат phyloxml и сохраняем
Phylo.write(tree, 'my_tree_data.phyloxml', 'phyloxml')
```

```python
# Задание 5. Проверяем, что все работает корректно, заново прочитав дерево
tree = Phylo.read('my_tree.phyloxml', 'phyloxml')
plt.rcParams['figure.figsize'] = [100, 100]
Phylo.draw(tree)
```

```python
### Задания на ETE toolkit ###
```

```python
# Задание 1. Читаем дерево и любуемся на него
target_url = 'https://www.jasondavies.com/tree-of-life/life.txt'
treedata = urlopen(target_url).read().decode('utf-8')
t = Tree(treedata, format=1)
ts = TreeStyle()
ts.show_leaf_name = True
ts.show_branch_support = True
t.show(tree_style=ts) # любуемся в красивом окошке
```

```python
print(t) # любуемся здесь
```

```python
# Задание 2. Обрезка деревьев
leaf_names_list = [] # создаем список листьев
for leaf in t:
  leaf_names_list.append(leaf.name)

loosers = random.sample(leaf_names_list, 42) # рандомно вытаскиваем оттуда 42 вида

t.prune(loosers)
```

```python
# Задание 3. Нарисуем обрезаное дерево
t.show(tree_style=ts)
```

```python
# Задание 4. Задание на поиграться с экстетиками
for n in t.traverse():
   nstyle = NodeStyle()
   nstyle["fgcolor"] = "red"
   nstyle["size"] = 5
   n.set_style(nstyle)
```

```python
t.show(tree_style=ts)
```
